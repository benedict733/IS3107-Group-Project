```python
import pickle
import pandas as pd
import numpy as np
user_mapping = pd.read_csv('user_mapping.csv', names = ['author', 'user_id_idx'])
game_mapping = pd.read_csv('game_mapping.csv',names = ['app_id', 'item_id_idx']) 
game_df = pd.read_csv('games.csv')
LightGCN_score_matrix = pickle.load(open(f'LightGCN_result.pkl', "rb"))
def recommend(author):
    result = {}
    # existing users
    if author in list(user_mapping['author']):
        print('collaborative filtering')
        user_id_idx = user_mapping.loc[user_mapping.author == author].user_id_idx.tolist()[0]
        # get corresponding row in the lightGCN scoring matrix
        game_score = LightGCN_score_matrix[user_id_idx]
        #indexes in descending orders:
        row_dec = (-1 * game_score).argsort()
        # get top 10 scoring games index
        top_10_index = row_dec[:10]

        # get their score
        top_10_score = row_dec[top_10_index].tolist()

        # match the top 10 scoring games index with app id
        top_10_games = game_mapping.loc[game_mapping.item_id_idx.isin(top_10_index.tolist())]['app_id'].tolist()

        game_name = game_df.loc[game_df.app_id.isin(top_10_games)].title.tolist()

        for i in range(10):
            result[(game_name[i],top_10_games[i])] = top_10_score[i]
    else:
        # content based -- return top 100
        # currently -- recommend newest releases
        print('content based')
        game_df_sorted = game_df.sort_values(by = 'date_release', ascending = False).head(10)[['app_id', 'title']].to_numpy()
        for i in range(10):
            result[tuple(game_df_sorted[i])] = 0
    return result

```

```python
for i in user_mapping.author.tolist()[:100]:
    print(i)
    print(recommend(i))
```
